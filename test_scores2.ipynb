```python
#rename and select necessary columns
test_df.rename(columns={"Name_post":"Names", "Score_pre":"Pre-Test Score", "Score_post":"Post-Test Score", "State_pre":"Location", "Geo-political zone_pre":"Zone"}, inplace=True)
test_df = test_df[["Email", "Names", "Pre-Test Score", "Post-Test Score", "Location", "Zone"]]
test_df
```

```python
#format the test scores
test_df["Pre-Test Score"] = test_df["Pre-Test Score"].str.split("/").str[0].str[:2]
test_df["Post-Test Score"] = test_df["Post-Test Score"].str.split("/").str[0].str[:2]
test_df
```

```python
#convert test scores to suitable datatype
test_df["Pre-Test Score"] = pd.to_numeric(test_df["Pre-Test Score"]).astype(int)
test_df["Post-Test Score"] = pd.to_numeric(test_df["Post-Test Score"]).astype(int)
```

```python
test_df["Difference"] = (test_df["Post-Test Score"]-test_df["Pre-Test Score"]).astype(int)
```

```python
test_df["Names"] = test_df["Names"].str.title()
```

```python
test_df.dtypes
```

```python
#create a new column "Remarks"
conditions = [
    ((test_df["Post-Test Score"]>=0) & (test_df["Post-Test Score"]<=17)),
    ((test_df["Post-Test Score"]>=18) & (test_df["Post-Test Score"]<=26)), 
    ((test_df["Post-Test Score"]>=27) & (test_df["Post-Test Score"]<=33)),
    ((test_df["Post-Test Score"]>=34))
]
values = ["Average", "Satisfactory", "Very good", "Excellent"]
test_df["Remark"] = np.select(conditions, values)
test_df
```

```python
test_df = test_df[["Names", "Location", "Pre-Test Score", "Post-Test Score", "Difference", "Remark", "Zone"]]
```
